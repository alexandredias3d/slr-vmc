## 2. Reading information related to the VM Placement Policy

Uses the utils module to load the information related to the VM placement policy as a dataframe.

```python
vmp_df = DataExtraction().get_vm_placement()
```

## 3. Ploting the bar charts
Firstly, we count the amount of times each method/criteria has been used by the papers.  
Then, we plot a horizontal bar chart with the absolute amounts and the relative amounts.  
Since a paper can use more than one method/criteria for the VM placement policy, the sum of the relative amounts can be greater than 100%.

### 3.1. Ploting the method bar chart

```python
data = count_method_occurrences(vmp_df, DataExtraction.VMP_TECHNIQUE)
data[DataExtraction.VMP_TECHNIQUE] = data[DataExtraction.VMP_TECHNIQUE].replace(Abbreviations.VMP)
plot_horizontal_bar_chart(data, 
                          x=Table.AMOUNT, y=DataExtraction.VMP_TECHNIQUE, 
                          xlim=(0, 75), 
                          filename='RQ3-VMP-Techniques.pdf', 
                          total=len(vmp_df))
```

### 3.2. Ploting the criteria bar chart

```python
data = count_criteria_occurrences(vmp_df, DataExtraction.VMP_CRITERIA)
data[DataExtraction.VMP_CRITERIA] = data[DataExtraction.VMP_CRITERIA].replace(Abbreviations.CRITERIA)
plot_horizontal_bar_chart(data.head(5), 
                          x=Table.AMOUNT, y=DataExtraction.VMP_CRITERIA, 
                          xlim=(0, 75), 
                          filename='RQ3-VMP-Criteria.pdf', 
                          total=len(vmp_df))
```

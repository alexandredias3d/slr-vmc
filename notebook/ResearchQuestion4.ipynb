## 1. Importing third party libraries and the Utils module

```python
from utils import *
```

## 2. Read information related to the experiment

Uses the utils module to load the information related to the experiment.

```python
experiment_df = DataExtraction().get_experiment()
```

## 3. Plot the bar charts
Firstly, we count the occurrences of type and platform of experiences that have been used by papers.  
Then, we plot horizontal bar charts for this data containing absolute and relative amounts.  

### 3.1. Plot the experiment type bar chart

```python
data = (experiment_df[DataExtraction.EXPERIMENT_TYPE]
                     .value_counts()
                     .reset_index()
                     .rename(columns={Table.INDEX: DataExtraction.EXPERIMENT_TYPE, DataExtraction.EXPERIMENT_TYPE: Table.AMOUNT}))

plot_horizontal_bar_chart(data,
                          x=Table.AMOUNT,
                          y=DataExtraction.EXPERIMENT_TYPE,
                          xlim=(0, 100),
                          total=len(experiment_df),
                          filename='RQ4-Type.pdf')
```

### 3.2. Plot the experiment platform bar chart
We count the amount of papers that use each criteria and show the complete list of criteria.  
Then, we plot a horizontal bar chart of the top 5 occurrences.

```python
data = (experiment_df[DataExtraction.EXPERIMENT_PLATFORM]
                      .value_counts()
                      .reset_index()
                      .rename(columns={Table.INDEX: DataExtraction.EXPERIMENT_PLATFORM, DataExtraction.EXPERIMENT_PLATFORM: Table.AMOUNT}))
                
plot_horizontal_bar_chart(data,
                          x=Table.AMOUNT,
                          y=DataExtraction.EXPERIMENT_PLATFORM,
                          xlim=(0, 100),
                          total=len(experiment_df),
                          filename='RQ4-Platform.pdf')
```

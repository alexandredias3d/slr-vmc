# Data Extraction Form - Research Question 6 (RQ6)
## What metrics are used to evaluate the VM consolidation solutions?

This notebook analyses the collected data to answer the RQ6.

```python
from utils import *
```

```python
metrics_df = DataExtraction().get_metrics()
```

```python
total_papers = len(metrics_df)
metrics = (metrics_df.str.split(',')
                     .explode()
                     .str.strip()
                     .value_counts()
                     .drop(''))


metrics = (metrics.reset_index()
                  .rename(columns={Table.INDEX: DataExtraction.METRICS, DataExtraction.METRICS: Table.AMOUNT}))

for index, columns in metrics.iterrows():
    print(f'{columns[0]:<50} {columns[1]:>2}')
```

```python
plot_horizontal_bar_chart(metrics.head(10), 
                          x=DataExtraction.AMOUNT, 
                          y=DataExtraction.METRICS, 
                          xlim=(0, 100), 
                          total=total_papers, 
                          filename='RQ6-Metrics.pdf')
```

## What techniques are used to select VMs from PMs to be migrated?

## 1. Importing third party libraries and the Utils module

```python
from utils import *
```

## 2. Reading information related to the VM Selection Policy

Uses the utils module to load the information related to the VM selection policy as a dataframe.

```python
vms_df = DataExtraction().get_vm_selection()
```

## 3. Ploting the bar charts

Firstly, we count the amount of times each method/criteria has been used by the papers.  
Then, we plot a horizontal bar chart with the absolute amounts and the relative amounts.  
Since a paper can use more than one method/criteria for the VM selection policy, the sum of the relative amounts can be greater than 100%.

### 3.1. Ploting the method bar chart

```python
data = count_method_occurrences(vms_df, DataExtraction.VMS_TECHNIQUE, abbreviation=True)
plot_horizontal_bar_chart(data.head(5), 
                          x=Table.AMOUNT, y=DataExtraction.VMS_TECHNIQUE, 
                          xlim=(0, 100), 
                          filename='RQ2-VMS-Techniques.pdf', 
                          total=len(vms_df))
```

```python
data
```

### 3.2. Ploting the criteria bar chart

```python
data = count_criteria_occurrences(vms_df, DataExtraction.VMS_CRITERIA)
data[DataExtraction.VMS_CRITERIA] = data[DataExtraction.VMS_CRITERIA].replace(Abbreviations.CRITERIA)
plot_horizontal_bar_chart(data.head(5), 
                          x=Table.AMOUNT, y=DataExtraction.VMS_CRITERIA, 
                          xlim=(0, 100), 
                          filename='RQ2-VMS-Criteria.pdf', 
                          total=len(vms_df))
```

```python
data
```

### 3.3. Ploting the amount of VMs selected bar chart

```python
data = (vms_df[DataExtraction.VMS_AMOUNT]
                                        .value_counts()
                                        .reset_index()
                                        .rename(columns={Table.INDEX: DataExtraction.VMS_AMOUNT, DataExtraction.VMS_AMOUNT: Table.AMOUNT}))
data[DataExtraction.VMS_AMOUNT] = data[DataExtraction.VMS_AMOUNT].replace(Abbreviations.VMS)
plot_horizontal_bar_chart(data, 
                          x=Table.AMOUNT, y=DataExtraction.VMS_AMOUNT, 
                          xlim=(0, 100), 
                          filename='RQ2-VMS-Amount.pdf',
                          total=len(vms_df))
```
